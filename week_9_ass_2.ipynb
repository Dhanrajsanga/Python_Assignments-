```python
# Matplotlib is a comprehensive library for creating static, interactive, and animated visualizations
# in Python. It is one of the most popular data visualization libraries in Python and is often used 
# in conjunction with NumPy and pandas for data analysis and plotting.

# Why is Matplotlib Used?

# Versatility: It can create a wide range of static, animated, and interactive plots.
# Ease of Use: Offers a MATLAB-like interface for users familiar with MATLAB, making it easier to 
# switch or integrate.
# Customization: Provides extensive customization options for creating intricate and professional-
# quality graphics.
# Integration: Works well with many operating systems and graphics backends, and integrates 
# seamlessly with other Python libraries like pandas and NumPy.
# Community Support: Being open-source and widely used, it has a strong community and 
# extensive documentation and resources.

# Five Types of Plots in Pyplot Module of Matplotlib:

# Line Plot: Useful for visualizing data trends over time. 
# It displays information as a series of data points called 'markers', connected by straight line segments.

# Bar Chart: Ideal for comparing multiple categories. These charts represent categorical data with 
# rectangular bars (vertical or horizontal) with lengths proportional to the values they represent.

# Histogram: Used for visualizing the distribution of a dataset. It groups data points into bins 
# and shows the frequency of occurrences in each bin.

# Scatter Plot: Displays values for two variables for a set of data. The data are displayed as
# a collection of points, each having the value of one variable determining the position
# on the horizontal axis and the value of the other variable determining the position on the vertical axis.

# Pie Chart: Represents data in a circular graph. The slices of the pie are proportional to the
# fraction of the whole for various categories.

# Matplotlib's Pyplot module provides a MATLAB-like interface and is typically used for simple
# and quick generation of plots, making it a favorite for exploratory data analysis.
```

```python
# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this 
# generated data plot a scatter plot.
```

```python
# A scatter plot is a type of data visualization that uses dots to represent the values
# obtained for two different variables - one plotted along the x-axis and the other plotted 
# along the y-axis. Scatter plots are used to observe and show relationships between two numeric
# variables. The pattern of the scatter plot can indicate the relationship between the variables,
# such as whether they might be positively or negatively correlated or whether no correlation exists
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for x and y
np.random.seed(3)  # This ensures the code produces the same results each time it's run
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create a scatter plot
plt.scatter(x, y)

# Adding title and labels
plt.title('Scatter Plot Example')
plt.xlabel('X Values')
plt.ylabel('Y Values')

# Show the plot
plt.show()
```

```python
# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
# Use the following data:
```

```python

# The subplot() function in Matplotlib is used to create multiple separate plots 
# (subplots) within the same figure. This is useful when you want to display several
# plots in a single window or image, enabling comparison or just compact presentation
# of different data sets.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for the plots
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Creating a figure containing 4 subplots
plt.figure(figsize=(10, 8))  # Optional: Specifies the size of the entire figure

# Plot 1
plt.subplot(2, 2, 1)  # (rows, columns, index)
plt.plot(x, y1, 'r')  # 'r' is the color red
plt.title('Plot 1')
plt.xlabel('X')
plt.ylabel('Y1')

# Plot 2
plt.subplot(2, 2, 2)
plt.plot(x, y2, 'g')  # 'g' is the color green
plt.title('Plot 2')
plt.xlabel('X')
plt.ylabel('Y2')

# Plot 3
plt.subplot(2, 2, 3)
plt.plot(x, y3, 'b')  # 'b' is the color blue
plt.title('Plot 3')
plt.xlabel('X')
plt.ylabel('Y3')

# Plot 4
plt.subplot(2, 2, 4)
plt.plot(x, y4, 'y')  # 'y' is the color yellow
plt.title('Plot 4')
plt.xlabel('X')
plt.ylabel('Y4')

# Display the plots
plt.tight_layout()  # Adjusts the subplots to fit into the figure area.
plt.show()
```

```python
# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontalbarplot.
# import numpy as np
# company = np.array(["Apple", "Microsoft", "Google", "AMD"])
# profit = np.array([3000, 8000, 1000, 10000])
```

```python
# A bar plot (or bar chart) is a graphical representation of data using bars of 
# different heights or lengths. Bar plots are used to compare different groups or 
# to track changes over time. The key characteristics of a bar plot are that it
# provides a visual comparison of discrete categories or groups, making it easier
# to interpret and understand categorical data.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Vertical Bar Plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit, color='blue')
plt.title('Profit by Company')
plt.xlabel('Company')
plt.ylabel('Profit in millions')
plt.show()

# Horizontal Bar Plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit, color='green')
plt.title('Profit by Company')
plt.xlabel('Profit in millions')
plt.ylabel('Company')
plt.show()
```

```python
# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
# box1 = np.random.normal(100, 10, 200)
# box2 = np.random.normal(90, 20, 200)
```

```python
# A box plot (also known as a box-and-whisker plot) is a standardized way of displaying the 
# distribution of data based on a five-number summary: minimum, first quartile (Q1), median, 
# third quartile (Q3), and maximum. It can tell you about your outliers and what their values 
# are. It can also tell you if your data is symmetrical, how tightly your data is grouped, 
# and if and how your data is skewed.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Creating the box plot
plt.figure(figsize=(8, 6))
plt.boxplot([box1, box2], labels=['Box1', 'Box2'])
plt.title('Box Plot Comparison')
plt.ylabel('Values')
plt.show()
```
