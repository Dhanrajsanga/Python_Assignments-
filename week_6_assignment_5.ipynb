```python
# Q1. Explain GET and POST methods.
```

```python
# GET Method:
# The GET method is one of the HTTP methods used by the World Wide Web. It is designed to request data from a specified resource.
# When a client sends a GET request to a server, it is asking the server to retrieve or return a resource.

# Common Use Cases:

# Retrieving information from a server.
# Navigating to a specific URL.
# Bookmarking a page.

# GET /search?query=Flask HTTP/1.1
# Host: www.example.com

# POST Method:
# The POST method is another HTTP method used for submitting data to be processed to a specified resource. 
# When a client sends a POST request to a server, it is asking the server to accept the data enclosed in the body of the request,
# often for the purpose of creating or updating a resource on the server.

# Common Use Cases:

# Submitting form data.
# Uploading a file.
# Creating a new resource on the server.

# POST /submit_form HTTP/1.1
# Host: www.example.com
# Content-Type: application/x-www-form-urlencoded

# username=johndoe&password=secret
```
