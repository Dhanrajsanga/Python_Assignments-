```python
# Q1. What is MongoDB? Explain non-relational databases in short. In which scenarios it is preferred to use
# MongoDB over SQL databases?
```

```python
# MongoDB is a popular, open-source NoSQL (non-relational) database management system. Unlike traditional relational databases, which organize data into tables with predefined 
# schemas, MongoDB stores data in flexible, JSON-like documents called BSON (Binary JSON) in a format similar to key-value pairs. It is designed to handle large amounts of 
# unstructured or semi-structured data and is particularly well-suited for applications with rapidly changing schemas.

# Non-relational databases, or NoSQL databases, represent a category of database systems that provide a flexible and scalable approach to storing and retrieving data. 
# They are characterized by their ability to handle diverse and dynamic data types, scalability across distributed architectures, and schema flexibility.

# In scenarios where data is dynamic, evolving, or doesn't fit neatly into a predefined schema, non-relational databases like MongoDB offer advantages. Some situations where
# MongoDB may be preferred over traditional SQL databases include:

# 1.Schema Flexibility: MongoDB allows for dynamic and flexible schemas, making it easier to accommodate changes in data structure without requiring a rigid, predefined schema.

# 2.Scalability: MongoDB is designed to scale horizontally, making it well-suited for applications with growing data volumes and increasing user loads. 
# It can distribute data across multiple servers and handle large amounts of traffic efficiently.

# 3.Complex Data: When dealing with complex, hierarchical, or nested data structures, MongoDB's document-oriented model can be more natural and efficient compared to the tabular 
# structure of relational databases.

# 4.Development Speed: In scenarios where rapid development and iteration are crucial, MongoDB's flexible document model enables developers to work quickly without the need for 
# extensive database schema planning.

# 5.Real-time Analytics: For applications that require real-time analytics and quick access to large volumes of data, MongoDB's indexing capabilities and efficient querying can be 
# advantageous.
```

```python
# Q2. State and Explain the features of MongoDB.
```

```python
# MongoDB is a feature-rich NoSQL database that offers various capabilities to meet the demands of modern applications. Here are some key features of MongoDB:

# 1.Document-Oriented:

# MongoDB stores data in flexible, JSON-like BSON documents, allowing for nested structures and arrays. Each document can have a different structure, providing schema flexibility.

# 2.Dynamic Schema:

# MongoDB doesn't enforce a rigid schema, allowing for changes to the data model without requiring modifications to existing data. This flexibility is beneficial in dynamic and 
# evolving application environments.

# 3.Indexing:

# MongoDB supports the creation of indexes on any field, improving query performance. It includes various types of indexes, such as single-field, compound, geospatial, and text indexes.

# 4.Aggregation Framework:

# MongoDB provides a powerful aggregation framework for data transformation and analysis. It supports a pipeline-based approach, allowing developers to perform complex data 
# manipulations and aggregations.

# 5.High Availability:

# MongoDB supports automatic sharding and replication, ensuring high availability and fault tolerance. Replication allows data to be mirrored across multiple nodes, and sharding 
# enables horizontal scaling by distributing data across multiple servers.

# 6.Horizontal Scalability:

# MongoDB is designed to scale horizontally, allowing for the distribution of data across multiple servers or clusters. This enables the database to handle large amounts of data 
# and traffic.

# 7.Geospatial Indexing:

# MongoDB includes support for geospatial indexing, making it suitable for location-based applications. This feature allows efficient querying and analysis of spatial data.

# 8.Document Validation:

# MongoDB provides the ability to enforce data validation rules at the database level. This ensures that data adheres to specific criteria, maintaining data integrity.

# 9.Atomic Operations:

# MongoDB supports atomic operations on a single document, allowing developers to perform complex updates and modifications in a way that ensures consistency.

# 10.Security Features:

# MongoDB includes various security features, such as authentication, authorization, and encryption. Access control mechanisms allow administrators to define fine-grained permissions
# for users and roles.

# 11.Schema Design Tools:

# MongoDB Atlas, the official cloud-based database service for MongoDB, offers schema design tools that help developers model their data effectively and optimize performance.

# 12.Text Search:

# MongoDB provides text search capabilities, allowing for efficient querying and indexing of textual data within documents.

# 13.Management Tools:

# MongoDB comes with management tools like MongoDB Compass, a graphical user interface for exploring and interacting with databases, making it easier for developers to 
# work with the database.
```

```python
# Q7. Explain why delete_one(), delete_many(), and drop() is used.
```

```python
# 1.delete_one():

# Purpose: Used to delete a single document that matches the specified criteria.
# Syntax:

# collection.delete_one(filter)
# Example:

# # Delete the document where the name is "Alice"
# collection.delete_one({"name": "Alice"})
# This method deletes only the first document that matches the given filter criteria.

# 2.delete_many():

# Purpose: Used to delete multiple documents that match the specified criteria.
# Syntax:

# collection.delete_many(filter)
# Example:

# # Delete all documents where the age is greater than 25
# collection.delete_many({"age": {"$gt": 25}})
# This method deletes all documents that match the given filter criteria.

# 3.drop():

# Purpose: Used to remove an entire collection from the database.
# Syntax:

# collection.drop()

# Example:

# # Drop (delete) the entire collection
# collection.drop()
```
