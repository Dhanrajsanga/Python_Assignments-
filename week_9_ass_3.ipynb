```python
# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
```

```python
# Seaborn is a popular data visualization library in Python, known for its ability to create aesthetically pleasing and informative statistical graphics.
# Here are five types of plots that you can create using Seaborn, along with their typical uses:

# Histogram:

# Use: A histogram represents the distribution of a univariate dataset. It shows the frequency of values in bins, giving an overview of the central 
# tendency, dispersion, and the shape of the distribution. Histograms are useful for understanding the underlying distribution of data, detecting
# outliers, and identifying different data patterns.
# Seaborn Function: sns.histplot()

# Scatter Plot:

# Use: Scatter plots are used for bivariate data analysis to show the relationship between two continuous variables. Each point represents an 
# observation in the dataset. This type of plot is useful for identifying correlations, trends, and outliers in data.
# Seaborn Function: sns.scatterplot()

# Box Plot:

# Use: A box plot (or box-and-whisker plot) displays the distribution of quantitative data and is particularly useful for indicating 
# whether a distribution is skewed and whether there are potential unusual observations (outliers) in the dataset. It shows the median, quartiles, and 
# outliers of the data.
# Seaborn Function: sns.boxplot()

# Heatmap:

# Use: Heatmaps are used to visualize matrix-like data and can be particularly effective in displaying correlations, clustering, or 
# the intensity of variables. They can help in identifying patterns, variations, and concentration of data based on color.
# Seaborn Function: sns.heatmap()

# Violin Plot:

# Use: Violin plots are similar to box plots but provide a deeper understanding of the density distribution of the data. 
# They are particularly useful for comparing the distribution of data across categories or groups. The thickness of the plot 
# at different values indicates the frequency or density of data points in that area.
# Seaborn Function: sns.violinplot()
# These plots are just a few examples of what Seaborn can do. 
# The library offers a wide range of plotting capabilities and is highly customizable, making it a preferred choice for data visualization in Python.
```

```python
# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
# "timepoint" and y = "signal" for different events and regions.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Create the line plot
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")

# Enhance the plot with title and labels (optional)
plt.title("FMRI Signal Across Time by Event and Region")
plt.xlabel("Timepoint")
plt.ylabel("Signal")

# Show the plot
plt.show()
```

```python
# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
# 'pclass', y = 'age' and y = 'fare'.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset
titanic = sns.load_dataset('titanic')

# Set up the matplotlib figure (optional, for better layout)
plt.figure(figsize=(14, 6))

# Create the first subplot for the 'age' distribution
plt.subplot(1, 2, 1)  # 1 row, 2 columns, first plot
sns.boxplot(x='pclass', y='age', data=titanic)
plt.title('Age Distribution by Passenger Class')

# Create the second subplot for the 'fare' distribution
plt.subplot(1, 2, 2)  # 1 row, 2 columns, second plot
sns.boxplot(x='pclass', y='fare', data=titanic)
plt.title('Fare Distribution by Passenger Class')

# Display the plots
plt.show()
```

```python
# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
# parameter for the 'cut' column of the diamonds dataset.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds = sns.load_dataset('diamonds')

# Set the size of the plot (optional, for better visualization)
plt.figure(figsize=(10, 6))

# Create the histogram
sns.histplot(data=diamonds, x='price', hue='cut', element='step', bins=30)

# Adding titles and labels (optional, for better readability)
plt.title('Distribution of Diamond Prices by Cut Quality')
plt.xlabel('Price')
plt.ylabel('Count')

# Display the plot
plt.show()
```

```python
# Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
# of the iris dataset.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset('iris')

# Create the pair plot
sns.pairplot(iris, hue='species')

# Display the plot
plt.show()
```

```python
# Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights = sns.load_dataset('flights')

# Reshape the dataset into a pivot table
flights_pivot = flights.pivot("month", "year", "passengers")

# Create the heatmap
plt.figure(figsize=(10, 8))  # Optional: Adjusts the size of the plot for better readability
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap='YlGnBu')  # 'annot' adds numeric labels on the cells

# Adding title (optional)
plt.title('Number of Passengers (in 1000s) Over the Years')

# Display the plot
plt.show()
```
