```python
# Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would
# use each type of test.
```

```python
# **Difference between t-test and z-test:**

# 1. **Sample Size:**
#    - **t-test:** Typically used when the sample size is small (usually less than 30) and the population
#     standard deviation is unknown.
#    - **z-test:** Used when the sample size is large (usually greater than 30) or when the population standard
# deviation is known.

# 2. **Standard Deviation:**
#    - **t-test:** Assumes that the population standard deviation is unknown and estimates it from the sample.
#    - **z-test:** Assumes that the population standard deviation is known.

# 3. **Distribution:**
#    - **t-test:** Uses the Student's t-distribution.
#    - **z-test:** Uses the standard normal distribution.

# **Example Scenarios:**

# 1. **Scenario for t-test:**
#    - **Situation:** Suppose you want to test whether the average test scores of a small sample of students
#     (less than 30) from a particular school are significantly different from the national average.
#    - **Test Type:** Use a one-sample t-test if you are comparing the sample mean to a known population mean.

# 2. **Scenario for z-test:**
#    - **Situation:** Suppose you have a large dataset of the heights of adult males in a country (more than 30),
#     and you want to test whether the average height is significantly different from a known population mean.
#    - **Test Type:** Use a z-test for one-sample or two-sample hypothesis testing when the sample size is large,
# or when the population standard deviation is known.

# In summary, choose a t-test when dealing with smaller sample sizes and unknown population standard deviations,
# while a z-test is appropriate for larger sample sizes or situations where the population standard deviation is
# known.
```

```python
# Q2: Differentiate between one-tailed and two-tailed tests.
```

```python
# **One-Tailed Test vs. Two-Tailed Test:**

# **1. Definition:**
#    - **One-Tailed Test:** In a one-tailed test, the critical region is located in only one tail of the 
#     distribution, either the right or the left.
#    - **Two-Tailed Test:** In a two-tailed test, the critical region is split between both tails of the 
# distribution.

# **2. Hypotheses:**
#    - **One-Tailed Test:** The null and alternative hypotheses are formulated in a way that specifies a 
#     directional relationship (greater than or less than).
#      - Example: \(H_0: \mu \leq 10\) and \(H_1: \mu > 10\) (right-tailed).
#    - **Two-Tailed Test:** The hypotheses are formulated to check if a parameter is different from a 
# certain value, without specifying the direction.
#      - Example: \(H_0: \mu = 10\) and \(H_1: \mu \neq 10\) (two-tailed).

# **3. Critical Region:**
#    - **One-Tailed Test:** Critical region is located in either the right tail or the left tail of the 
#     distribution.
#    - **Two-Tailed Test:** Critical region is split between both tails of the distribution.

# **4. Rejection Area:**
#    - **One-Tailed Test:** Rejection occurs only in one direction (either greater than or less than).
#    - **Two-Tailed Test:** Rejection can occur in either direction (greater than or less than), depending
# on the observed result falling in the critical region.

# **5. Example:**
#    - **One-Tailed Test:** Testing whether a new drug increases blood pressure significantly
#     (one-tailed as we are interested if it increases, not decreases).
#    - **Two-Tailed Test:** Testing whether there is a difference in mean scores before and after treatment.

# **6. Directional vs. Non-Directional:**
#    - **One-Tailed Test:** Directional (specifying the direction of the effect).
#    - **Two-Tailed Test:** Non-directional (checking for any significant difference).

# In summary, the choice between one-tailed and two-tailed tests depends on the specific hypothesis being
# tested and the directional nature of the research question. One-tailed tests are used when researchers 
# are interested in a specific direction of the effect, while two-tailed tests are more conservative and
# used when researchers want to detect any significant difference, regardless of direction.
```

```python
# Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
# each type of error.
```

```python
# **Type I and Type II Errors in Hypothesis Testing:**

# **1. Type I Error (False Positive):**
#    - **Definition:** Occurs when you reject a null hypothesis that is actually true.
#     It is the probability of concluding that there is an effect when there is no effect.
#    - **Symbolized as:** \(\alpha\) (alpha).
#    - **Example Scenario:** Conducting a drug trial, where the null hypothesis (\(H_0\)) is
#     that the drug has no effect. A Type I error would occur if the trial incorrectly concludes
#     that the drug is effective when, in reality, it has no effect.

# **2. Type II Error (False Negative):**
#    - **Definition:** Occurs when you fail to reject a null hypothesis that is actually false.
#     It is the probability of concluding that there is no effect when there is an effect.
#    - **Symbolized as:** \(\beta\) (beta).
#    - **Example Scenario:** Continuing with the drug trial example, a Type II error would occur
#     if the trial incorrectly fails to detect the effectiveness of the drug, leading to the conclusion
#     that the drug has no effect when it actually does.

# **Illustrative Scenario:**
#    - **Context:** Suppose a company claims that a new manufacturing process improves the strength of its products.
#    - **Hypotheses:**
#      - \(H_0\): The new process has no effect (\(H_0: \mu = 0\)).
#      - \(H_1\): The new process improves strength (\(H_1: \mu > 0\)).
#    - **Possible Outcomes:**
#      - **Outcome 1 (Reality):** The new process does improve strength (\(H_1\) is true).
#        - **Type I Error (False Positive):** Concluding that the process improves strength when, 
#         in fact, it does not (\(\alpha\)).
#        - **Correct Decision:** Concluding that the process has no effect (Correct decision).
#      - **Outcome 2 (Reality):** The new process has no effect (\(H_0\) is true).
#        - **Correct Decision:** Concluding that the process has no effect (Correct decision).
#        - **Type II Error (False Negative):** Failing to detect the improvement when it exists (\(\beta\)).

# **Summary:**
#    - **Type I Error:** Incorrectly rejecting a true null hypothesis.
#    - **Type II Error:** Incorrectly failing to reject a false null hypothesis.

# Understanding Type I and Type II errors is crucial for researchers and decision-makers, as it helps 
# in evaluating the risks associated with hypothesis testing and making informed decisions based on the results.
```

```python
# Q4: Explain Bayes's theorem with an example.
```

```python
# **Bayes's Theorem:**

# Bayes's Theorem is a mathematical formula that describes the probability of an event based on 
# prior knowledge of conditions that might be related to the event. It is named after the Reverend 
# Thomas Bayes and is fundamental in Bayesian statistics.

# The formula is given by:

# \[ P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)} \]

# where:
# - \( P(A|B) \) is the probability of event A given that event B has occurred.
# - \( P(B|A) \) is the probability of event B given that event A has occurred.
# - \( P(A) \) and \( P(B) \) are the marginal probabilities of events A and B, respectively.

# **Example: Medical Diagnosis**

# Suppose we have a medical test for a rare disease, and the prevalence of the disease in the population 
# is low (let's say 1%). The test has a sensitivity of 95% (correctly identifies individuals with the disease) 
#         and a specificity of 90% (correctly identifies individuals without the disease).

# Now, let's calculate the probability of having the disease given a positive test result using Bayes's
#         Theorem.

# - \( A \): Having the disease
# - \( B \): Positive test result

# 1. **Given Data:**
#    - \( P(A) \): Prevalence of the disease = 0.01 (1%)
#    - \( P(B|A) \): Sensitivity of the test = 0.95 (95%)
#    - \( P(B|\neg A) \): False positive rate = 0.10 (10%)

# 2. **Calculate \( P(A|B) \):**
#    \[ P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)} \]

#    \[ P(A|B) = \frac{(0.95 \cdot 0.01)}{(0.95 \cdot 0.01) + (0.10 \cdot 0.99)} \]

#    \[ P(A|B) \approx \frac{0.0095}{0.0095 + 0.099} \approx \frac{0.0095}{0.1085} \approx 0.0878 \]

# Therefore, given a positive test result, the probability of actually having the disease 
#         is approximately 8.78%. Despite the positive test result, the low prevalence of 
#         the disease and the false positive rate of the test contribute to a relatively 
#         low probability of truly having the disease. This illustrates the impact of prior 
#         probabilities on the interpretation of test results.
```

```python
# Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.
```

```python
# **Confidence Interval:**

# A confidence interval is a range of values that is used to estimate the true value of a population parameter 
# with a certain level of confidence. It provides a range within which we expect the true parameter to lie
# based on the information obtained from a sample.

# **Calculation of Confidence Interval:**

# The formula for calculating a confidence interval for a population mean (\(\mu\)) is given by:

# \[ \text{Confidence Interval} = \bar{X} \pm Z \left( \frac{s}{\sqrt{n}} \right) \]

# where:
# - \(\bar{X}\) is the sample mean.
# - \(Z\) is the critical value from the standard normal distribution, corresponding to the desired confidence
# level.
# - \(s\) is the sample standard deviation.
# - \(n\) is the sample size.

# **Example: Confidence Interval for Mean Height**

# Suppose we want to estimate the average height of adult females in a population. We collect a 
# random sample of 50 females and measure their heights. The sample mean height is 65 inches, and 
# the sample standard deviation is 3 inches.

# 1. **Select Confidence Level:**
#    - Let's choose a 95% confidence level (\(Z = 1.96\) for a two-tailed test).

# 2. **Plug in Values:**
#    - Sample mean (\(\bar{X}\)): 65 inches
#    - Sample standard deviation (\(s\)): 3 inches
#    - Sample size (\(n\)): 50
#    - Critical value (\(Z\)): 1.96 (for 95% confidence)

# 3. **Calculate Confidence Interval:**
#    \[ \text{Confidence Interval} = 65 \pm 1.96 \left( \frac{3}{\sqrt{50}} \right) \]

# 4. **Result:**
#    - \[ \text{Confidence Interval} \approx 65 \pm 1.96 \times \left( \frac{3}{\sqrt{50}} \right) \]
#    - \[ \text{Confidence Interval} \approx 65 \pm 0.843 \]
#    - \[ \text{Confidence Interval} \approx (64.157, 65.843) \]

# Therefore, we can say with 95% confidence that the true average height of adult females in the 
# population is estimated to be between 64.157 inches and 65.843 inches based on our sample data.
```

```python
# Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the
# event's probability and new evidence. Provide a sample problem and solution.
```

```python
# **Bayes' Theorem Example: Probability of Disease Given Test Result**

# Suppose there is a medical test for a certain disease. The prevalence of the disease in the population 
# is 2%, and the test has a sensitivity of 90% and a specificity of 85%.

# **Given Data:**
# - Prevalence of the disease (\(P(D)\)): 0.02 (2%)
# - Sensitivity of the test (\(P(\text{Positive Test | Disease})\)): 0.90 (90%)
# - Specificity of the test (\(P(\text{Negative Test | No Disease})\)): 0.85 (85%)

# **Problem:**
# If a randomly selected individual tests positive for the disease, what is the probability that they
# actually have the disease?

# **Solution using Bayes' Theorem:**
# Let:
# - \(D\): The event of having the disease.
# - \(T^+\): The event of testing positive for the disease.

# We want to find \(P(D | T^+)\), the probability of having the disease given a positive test result.

# Bayes' Theorem is given by:
# \[ P(D | T^+) = \frac{P(T^+ | D) \cdot P(D)}{P(T^+)} \]

# 1. **Calculate \(P(T^+)\):**
# \[ P(T^+) = P(T^+ | D) \cdot P(D) + P(T^+ | \neg D) \cdot P(\neg D) \]

#    \[ P(T^+) = (0.90 \cdot 0.02) + (0.15 \cdot 0.98) \]
#    \[ P(T^+) = 0.018 + 0.147 \]
#    \[ P(T^+) = 0.165 \]

# 2. **Apply Bayes' Theorem:**
# \[ P(D | T^+) = \frac{P(T^+ | D) \cdot P(D)}{P(T^+)} \]

#    \[ P(D | T^+) = \frac{(0.90 \cdot 0.02)}{0.165} \]
#    \[ P(D | T^+) = \frac{0.018}{0.165} \]
#    \[ P(D | T^+) \approx 0.109 \]

# **Conclusion:**
# If an individual tests positive for the disease, the probability that they actually have the disease 
# is approximately 10.9%. Despite the positive test result, the low prevalence of the disease in the 
# population contributes to a relatively low probability of truly having the disease. 
# illustrates the importance of considering both sensitivity and specificity in the 
# interpretation of medical test results.
```

```python
# Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error?
# Provide an example of a scenario where a larger sample size would result in a smaller margin of error.
```

```python
# **Margin of Error in a Confidence Interval:**

# The margin of error in a confidence interval is the range around the sample estimate within which we expect the true population parameter 
# to lie with a certain level of confidence. It represents the precision of our estimate and is influenced by factors such as the sample size 
# and the desired level of confidence.

# **Formula for Margin of Error:**
# \[ \text{Margin of Error} = Z \times \left( \frac{s}{\sqrt{n}} \right) \]

# Where:
# - \( Z \) is the critical value from the standard normal distribution based on the desired confidence level.
# - \( s \) is the sample standard deviation.
# - \( n \) is the sample size.

# **Effect of Sample Size on Margin of Error:**

# 1. **Inverse Relationship:** The margin of error is inversely proportional to the square root of the sample size (\( \sqrt{n} \)). 
# As the sample size increases, the square root of the sample size increases, leading to a decrease in the margin of error.

# 2. **Larger Sample Size, Smaller Margin of Error:** A larger sample size generally results in a smaller margin of error, indicating 
# a more precise estimate of the population parameter.

# **Example Scenario:**

# Suppose you are conducting a survey to estimate the average income of households in a city. If you take a small sample of 50 households, 
# the margin of error might be relatively large, resulting in a wider confidence interval. However, if you increase the sample size to 500 
# households while keeping other factors constant, the margin of error would decrease, leading to a narrower confidence interval.
# A larger sample size allows for a more precise estimation of the average income, reducing uncertainty and increasing the reliability of the results.
```

```python
# Q9. Calculate the z-score for a data point with a value of 75, a population mean of 70, and a population
# standard deviation of 5. Interpret the results.
```

```python
# To calculate the z-score for a data point, you can use the following formula:

# \[ Z = \frac{X - \mu}{\sigma} \]

# Where:
# - \( X \) is the data point's value.
# - \( \mu \) is the population mean.
# - \( \sigma \) is the population standard deviation.

# In this case:
# - \( X = 75 \) (data point's value)
# - \( \mu = 70 \) (population mean)
# - \( \sigma = 5 \) (population standard deviation)

# **Calculation:**
# \[ Z = \frac{75 - 70}{5} = 1 \]

# **Interpretation:**
# A z-score of 1 indicates that the data point is one standard deviation above the population mean. 
#                 In other words, the value of 75 is one standard deviation away from the mean of 70 in the positive direction. 
#                 Z-scores help to standardize values, allowing for comparisons between data points from different normal distributions.
```

```python
# Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average
# of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is
# significantly effective at a 95% confidence level using a t-test.
```

```python
# To conduct a hypothesis test for the effectiveness of the new weight loss drug, we'll use a t-test for the population mean.
# The null hypothesis (\(H_0\)) typically assumes no effect, while the alternative hypothesis (\(H_1\)) asserts the presence of an effect
# . The t-test formula is:

# \[ t = \frac{\bar{X} - \mu_0}{\frac{s}{\sqrt{n}}} \]

# where:
# - \(\bar{X}\) is the sample mean,
# - \(\mu_0\) is the hypothesized population mean under the null hypothesis,
# - \(s\) is the sample standard deviation,
# - \(n\) is the sample size.

# In this case:
# - Sample mean (\(\bar{X}\)) = 6 pounds,
# - Hypothesized population mean (\(\mu_0\)) = 0 (assuming no weight loss),
# - Sample standard deviation (\(s\)) = 2.5 pounds,
# - Sample size (\(n\)) = 50.

# **Null Hypothesis (\(H_0\)):**
# \[ H_0: \mu = \mu_0 \]
# \[ H_0: \text{The new drug has no effect on weight loss (\(\mu = 0\)).} \]

# **Alternative Hypothesis (\(H_1\)):**
# \[ H_1: \mu \neq \mu_0 \]
# \[ H_1: \text{The new drug is significantly effective in weight loss.} \]

# **Significance Level (\(\alpha\)):** 0.05 (95% confidence level).

# **Degrees of Freedom (\(df\)):** \(n - 1 = 50 - 1 = 49\).

# **Critical Value for Two-Tailed Test:** \(t_{\alpha/2, df}\).

# Now, calculate the t-statistic:
# \[ t = \frac{6 - 0}{\frac{2.5}{\sqrt{50}}} \]

# **Calculations:**
# \[ t \approx \frac{6 - 0}{\frac{2.5}{\sqrt{50}}} \]
# \[ t \approx \frac{6}{\frac{2.5}{\sqrt{50}}} \]
# \[ t \approx \frac{6}{\frac{2.5}{\sqrt{50}}} \approx 10.78 \]

# Compare the calculated t-statistic with the critical value for a two-tailed test at \( \alpha = 0.05 \) and \( df = 49 \).

# If \( |t| > t_{\alpha/2, df} \), reject the null hypothesis.

# **Conclusion:**
# Since the calculated t-statistic (10.78) is much larger than the critical value, we reject the null hypothesis. 
# There is sufficient evidence to conclude that the new weight loss drug is significantly effective in promoting weight loss at a 95% confidence level.
```

```python
# Q11. In a survey of 500 people, 65% reported being satisfied with their current job. Calculate the 95%
# confidence interval for the true proportion of people who are satisfied with their job.
```

```python
# Q12. A researcher is testing the effectiveness of two different teaching methods on student performance.
# Sample A has a mean score of 85 with a standard deviation of 6, while sample B has a mean score of 82
# with a standard deviation of 5. Conduct a hypothesis test to determine if the two teaching methods have a
# significant difference in student performance using a t-test with a significance level of 0.01.
```

```python
# To determine if there is a significant difference in student performance between the two teaching methods, we can conduct a hypothesis
# test for the difference in means using a t-test. The formula for the t-statistic for independent samples is:

# \[ t = \frac{(\bar{X}_1 - \bar{X}_2)}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}} \]

# Where:
# - \(\bar{X}_1\) and \(\bar{X}_2\) are the sample means.
# - \(s_1\) and \(s_2\) are the sample standard deviations.
# - \(n_1\) and \(n_2\) are the sample sizes.

# Given:
# - \(\bar{X}_1 = 85\), \(s_1 = 6\), \(n_1\) (not provided)
# - \(\bar{X}_2 = 82\), \(s_2 = 5\), \(n_2\) (not provided)
# - Significance level (\(\alpha\)) = 0.01

# **Null Hypothesis (\(H_0\)):**
# \[ H_0: \mu_1 = \mu_2 \]
# \[ H_0: \text{There is no significant difference in student performance between the two teaching methods.} \]

# **Alternative Hypothesis (\(H_1\)):**
# \[ H_1: \mu_1 \neq \mu_2 \]
# \[ H_1: \text{There is a significant difference in student performance between the two teaching methods.} \]

# **Significance Level (\(\alpha\)):** 0.01

# **Degrees of Freedom (\(df\)):** \(df = n_1 + n_2 - 2\)

# **Critical Value for Two-Tailed Test:** \(t_{\alpha/2, df}\)

# Now, we need to calculate the t-statistic and compare it with the critical value for a two-tailed test.

# **Calculations:**
# \[ t = \frac{(85 - 82)}{\sqrt{\frac{6^2}{n_1} + \frac{5^2}{n_2}}} \]

# The degrees of freedom (\(df\)) depend on the sample sizes \(n_1\) and \(n_2\), which are not provided in the question. 
# You'll need to know these values to complete the calculations and make a conclusion.

# **Conclusion:**
# 1. If \(|t| > t_{\alpha/2, df}\), reject the null hypothesis.
# 2. If \(|t| \leq t_{\alpha/2, df}\), fail to reject the null hypothesis.

# You can use statistical software or a t-table to find the critical value based on the degrees of freedom and significance level to make a decision.
```

```python
# Q13. A population has a mean of 60 and a standard deviation of 8. A sample of 50 observations has a mean
# of 65. Calculate the 90% confidence interval for the true population mean.
```

```python
# To calculate the 90% confidence interval for the true population mean, we'll use the formula for the confidence interval for a population mean:

# \[ \text{Confidence Interval} = \bar{X} \pm Z \left( \frac{\sigma}{\sqrt{n}} \right) \]

# where:
# - \(\bar{X}\) is the sample mean,
# - \(Z\) is the critical value from the standard normal distribution based on the desired confidence level,
# - \(\sigma\) is the population standard deviation,
# - \(n\) is the sample size.

# Given:
# - Sample mean (\(\bar{X}\)) = 65,
# - Population mean (\(\mu\)) is not provided (assuming it is 60),
# - Population standard deviation (\(\sigma\)) = 8,
# - Sample size (\(n\)) = 50,
# - Confidence level = 90%.

# **Critical Value for 90% Confidence Level:** \(Z \approx 1.645\) (standard normal distribution).

# **Calculation:**
# \[ \text{Confidence Interval} = 65 \pm 1.645 \left( \frac{8}{\sqrt{50}} \right) \]

# **Interpretation:**
# The 90% confidence interval for the true population mean is calculated as follows:

# \[ \text{Confidence Interval} = 65 \pm 1.645 \left( \frac{8}{\sqrt{50}} \right) \]

# Now calculate the values:
# \[ \text{Confidence Interval} = 65 \pm 1.645 \times \left( \frac{8}{\sqrt{50}} \right) \]

# \[ \text{Confidence Interval} = 65 \pm 1.645 \times \left( \frac{8}{\sqrt{50}} \right) \approx (63.59, 66.41) \]

# Therefore, we are 90% confident that the true population mean falls between approximately 63.59 and 66.41 based on the sample data.
```
